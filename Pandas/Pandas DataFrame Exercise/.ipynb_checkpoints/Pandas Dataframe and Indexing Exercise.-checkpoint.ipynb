### DataFrame Methods :-
    1) Value_counts() :-Count the frequency of item.(Series/Dataframe)
    2) sort_values ():- (Series/Dataframe)
    3) Rank :- It provide the rank based on values.
    4) Sort_index() (Series/DataFrame).
    5) set_index()(DataFrame) :- Set our dataframe columns as index.
    5) reset_index()(Series/DataFrame):- Convert our index columns as normal column.
    6) Rename :- Renaming the existing column/index name.
    7) unique :- Gives the all unique value in series (Series).It count nan values
    8) nunique()(Series/Dataframe) :- it gives the numbr of unique values.It doesn't count nan values.
    9) isnull()/notnull()
    10) dropna()(Series/Dataframe) :- to drop the nan values.
    11) fillna()(Series.Dataframe) :-
    12) drop_duplicates():- dropping duplicate data.
    13) Drop():- it help to delete rows and columns from series or Dataframe.
    14) apply() :- It is used to apply custom function on series and dataframe.
